```python
#bibliotecas utilizadas
import numpy as np
import pandas as pd
```

```python
obt = pd.read_csv("C:/Users/Flavia Manhães/OneDrive - Universidade Federal de Pernambuco/2021.2/CONDEPE FIDEM/BDE II/ELIANE - DADOS/SAUDE/2019/A100614189_28_143_208.csv",sep=';', encoding='latin1',  engine='python')
nasc = pd.read_csv("C:/Users/Flavia Manhães/OneDrive - Universidade Federal de Pernambuco/2021.2/CONDEPE FIDEM/BDE II/ELIANE - DADOS/SAUDE/2019/A100634189_28_143_208.csv",sep=';', encoding='latin1',  engine='python')

def ajeitandoBanco(df,coluna):
    # Descartando colunas com valor nulo para evitar erros
    df.dropna(inplace = True)

    # Nova Data com Split das coluna "Name" separado por espaços
    new = coluna.str.split(" ", n = 1, expand = True)

    # Criando a Nova Coluna "First Name" com o new[0]
    df["Geocodigo"]= new[0] 

    # Criando a Nova Coluna "Last Name" com o new[1]
    df["Municipio"]= new[1] 

    # Retirando a antiga coluna "Name" 
    df.drop(columns =["Município"], inplace = True) 

# Imprimindo Alteração no Console
ajeitandoBanco(obt,obt["Município"])
ajeitandoBanco(nasc,nasc["Município"])
```

```python
# Merge utilizando a coluna do geocodigo
mortInfantil = pd.merge(obt, nasc, on=["Geocodigo","Municipio"])
mortInfantil["Taxa de mortalidade infantil"]=((mortInfantil["Óbitos_p/Residênc"]*1000)/mortInfantil["Nascim_p/resid.mãe"])
mortInfantil
```

```python
# Reposicionar colunas do dataframe

cols = mortInfantil.columns.tolist()
cols_=[cols[2],cols[3],cols[0],cols[4]]
mortInfantil=mortInfantil[cols_] 
mortInfantil
```

```python
# Renomear Colunas no Dataframe
mortInfantil.rename(columns={'Municipio': 'Municípios', "Nascim_p/resid.mãe": "Nascidos vivos", \
    "Óbitos_p/Residênc": "Óbito infantil"}, inplace=True)##Renomear colunas
```
